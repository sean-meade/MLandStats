# Two variables in pyplot


Adapted from: https://matplotlib.org/tutorials/introductory/pyplot.html

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r.')
```

```python
x = np.linspace(0.0, 100.0, 10)
```

$$ y = 2 x $$

```python
y1 = x * x
y2 = x * x * x
y3 = x * x * x * x

plt.plot(x, y1, 'r.')
plt.plot(x, y2, 'b.')
plt.plot(x, y3, 'y')
```

```python
z = np.linspace(0.0, 100.0, 11)
a1 = 2.0 * z + 5.0
a2 = 1.0 * z + 5.0
a3 = -2.0 * z + 0.0

plt.plot(z, a1, 'r')
plt.plot(z, a2, 'g')
plt.plot(z, a3, 'b')
```
